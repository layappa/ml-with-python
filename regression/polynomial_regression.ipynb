```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import root_mean_squared_error
```

```python
df = pd.read_csv(r'R:\aiml\Machine_Learning\regression\positional_datasets.csv')

print(df.head())
```

```python
x=df.iloc[:, 1:-1]
y=df.iloc[:,-1]
```

```python
model=LinearRegression()
model.fit(x,y)
```

```python
plt.scatter(x,y,color="red")
plt.plot(x,model.predict(x),color="blue")
plt.title("linear regression")
plt.xlabel("position")
plt.ylabel("salery")
plt.show()
```

```python
poly_model=PolynomialFeatures(degree=5)
x_poly=poly_model.fit_transform(x)
new_model=LinearRegression()
new_model.fit(x_poly,y)
```

```python
plt.scatter(x,y,color="red")
plt.plot(x,new_model.predict(x_poly),color="blue")
plt.title("polynomial regression")
plt.xlabel("salery")
plt.ylabel("position")
plt.show()
```

```python
a=(root_mean_squared_error(y,new_model.predict(x_poly)))
```

```python
c=(root_mean_squared_error(y,new_model.predict(x_poly))/(df['Salary'].mean()) ) *100
c
```
