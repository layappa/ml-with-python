using mode

```python
dataset["Gender"].fillna(dataset["Gender"].mode()[0],inplace=True)
```

```python
for i in dataset.select_dtypes(include="object").columns :
    dataset[i].fillna(dataset[i].mode()[0],inplace=True)
```

#handling missing value for numerical data types


```python
dataset.select_dtypes(include="float64").columns
```

```python
from sklearn.impute import SimpleImputer
```

```python
si=SimpleImputer(strategy="mean")
ar=si.fit_transform(dataset[['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount',
       'Loan_Amount_Term', 'Credit_History']])
ar

```

```python
new_dataset=pd.DataFrame(ar,columns=dataset.select_dtypes(include="float64").columns)
```
